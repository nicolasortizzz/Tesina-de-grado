```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import seaborn as sns
import numpy as np
import matplotlib.font_manager as fm

# Verificar si Rockwell está disponible
available_fonts = [f.name for f in fm.fontManager.ttflist]
rockwell_font = "Rockwell" if "Rockwell" in available_fonts else "serif"

# Crear figura
fig, ax = plt.subplots(figsize=(6, 6))
sns.set(style="whitegrid")

# Dibujar rectángulos en las esquinas
colors = ['#e0f7fa'] * 4
positions = [(0, 1), (1, 1), (0, 0), (1, 0)]  # Estanflación, Recalentamiento, Reflación, Recuperación
labels = ['Estanflación', 'Recalentamiento', 'Reflación', 'Recuperación']

for pos, color, label in zip(positions, colors, labels):
    rect = patches.Rectangle((pos[0], pos[1]), 1, 1, linewidth=2,
                             edgecolor='black', facecolor=color)
    ax.add_patch(rect)
    ax.text(pos[0] + 0.5, pos[1] + 0.5, label,
            ha='center', va='center', fontsize=14,
            fontweight='bold', fontname=rockwell_font)

# Límites y ejes
ax.set_xlim(0, 2.2)
ax.set_ylim(0, 2.2)
ax.set_xticks([])
ax.set_yticks([])

# Flechas eje crecimiento e inflación
ax.annotate('', xy=(1.95, 2.15), xytext=(0.05, 2.15),
            arrowprops=dict(facecolor='blue', edgecolor='blue', linewidth=2, headwidth=10))
ax.text(1.0, 2.18, 'Crecimiento', ha='center', va='bottom', fontsize=12,
        fontweight='bold', fontname=rockwell_font, color='black')

ax.annotate('', xy=(2.15, 1.95), xytext=(2.15, 0.05),
            arrowprops=dict(facecolor='blue', edgecolor='blue', linewidth=2, headwidth=10))
ax.text(2.18, 1.0, 'Inflación', ha='left', va='center', fontsize=12,
        fontweight='bold', fontname=rockwell_font, rotation=90, color='black')

# Flechas curvas en sentido horario más grandes
arrow_color = '#00acc1'
arrowstyle = patches.ArrowStyle.CurveFilledB(head_length=3.5, head_width=5.0)

# Aumentamos el radio del círculo
arrow1 = patches.FancyArrowPatch((0.7, 1.0), (1.0, 0.7),
                                  connectionstyle="arc3,rad=0.85",
                                  arrowstyle=arrowstyle,
                                  color=arrow_color, linewidth=5)
arrow2 = patches.FancyArrowPatch((1.0, 0.7), (1.3, 1.0),
                                  connectionstyle="arc3,rad=0.85",
                                  arrowstyle=arrowstyle,
                                  color=arrow_color, linewidth=5)
arrow3 = patches.FancyArrowPatch((1.3, 1.0), (1.0, 1.3),
                                  connectionstyle="arc3,rad=0.85",
                                  arrowstyle=arrowstyle,
                                  color=arrow_color, linewidth=5)
arrow4 = patches.FancyArrowPatch((1.0, 1.3), (0.7, 1.0),
                                  connectionstyle="arc3,rad=0.85",
                                  arrowstyle=arrowstyle,
                                  color=arrow_color, linewidth=5)

for arrow in [arrow1, arrow2, arrow3, arrow4]:
    ax.add_patch(arrow)

# Eliminar bordes
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Coordenadas ajustadas según pedido
points = {
    'A': (0.3, 1.0),
    'B': (0.3, 2.0),
    'C': (0.2, 1.3),
    'D': (0.6, 1.3)
}

fig, ax = plt.subplots()

# Dibujar puntos y etiquetas
for label, (x, y) in points.items():
    ax.plot(x, y, 'ko')  # punto negro
    ax.text(x + 0.01, y + 0.01, label, fontsize=12)

# Línea vertical de A a B
ax.plot([points['A'][0], points['B'][0]], [points['A'][1], points['B'][1]], 'k-', linewidth=0.5)

# Línea horizontal de C a D
ax.plot([points['C'][0], points['D'][0]], [points['C'][1], points['D'][1]], 'k-', linewidth=0.5)

# Etiquetas de los ejes
ax.set_xlabel(r'$\sigma_P$', fontsize=14)
ax.set_ylabel(r'$\mu_P$', fontsize=14)

# Quitar ticks
ax.set_xticks([])
ax.set_yticks([])

# Límites ajustados
ax.set_xlim(0.1, 0.7)
ax.set_ylim(0.9, 2.1)

# Quitar bordes innecesarios
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Desactivar marcas y cuadrícula
ax.tick_params(left=False, bottom=False)
ax.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos de la tabla, sin asteriscos
data = [
    ["Petróleo y Gas",     "99.8%", "99.2%", "95.8%", "95.0%", "94.4%", "91.9%"],
    ["Farmacéutico",       "99.7%", "99.3%", "92.6%", "91.4%", "90.3%", "-"],
    ["Sector industrial",  "98.7%", "97.2%", "-",     "-",     "-",     "-"],
    ["Materiales básicos", "97.8%", "91.4%", "-",     "-",     "-",     "-"],
    ["Consumo básico",     "97.0%", "91.3%", "-",     "-",     "-",     "-"],
    ["Sector financiero",  "96.9%", "92.3%", "-",     "-",     "-",     "-"]
]

columns = ["ESTANFLACIÓN", "Consumo Discrecional", "Tecnología", "Sector Industrial",
           "Sector Financiero", "Telecomunicaciones", "Materiales Básicos"]

# Crear DataFrame
df = pd.DataFrame(data, columns=columns)

# Crear figura y eje
fig, ax = plt.subplots(figsize=(14, 5))
ax.axis('off')

# Crear la tabla
table = ax.table(cellText=df.values,
                 colLabels=df.columns,
                 cellLoc='center',
                 loc='center',
                 colColours=["#d2e9f7"] + ["white"] * (len(columns) - 1))

# Estilo de la tabla
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.2, 2)

# Resaltar primera columna y fila
for (row, col), cell in table.get_celld().items():
    if row == 0 or col == 0:
        cell.set_facecolor("#d2e9f7")  # celeste claro
        cell.set_text_props(weight='bold')

# Ajuste de ancho de columnas
for col in range(len(columns)):
    table.auto_set_column_width(col)

# Guardar imagen
image_path = "/mnt/data/tabla_tesis_estanflacion_ajustada.png"
plt.savefig(image_path, bbox_inches='tight', dpi=300)

image_path
```
